```python
# Define input and output variables
inputs = data.iloc[:, 1:4]  # Use first 3 columns as inputs
outputs = data.iloc[:, 4:]  # Use remaining columns as outputs
```

```python
prob = pydea.DEAProblem(inputs, outputs, returns='CRS')
myresults = prob.solve()

# Print results
print("Status:")
print(myresults['Status'])
print("Efficiency scores:")
print(myresults['Efficiency'])
print("Weights:")
print(myresults['Weights'])
# print(dea_model.input_weights)
# print("Weights for outputs:")
# print(dea_model.output_weights)
```

```python
weight_df = pd.DataFrame(myresults['Weights'])
```

```python
weight_df
```

```python
efficiency_df = pd.DataFrame(myresults["Efficiency"])
efficiency_df.index = data["dmu"]
```

```python
efficiency_df
```

```python
# Sort DataFrame by efficiency scores in descending order
df_sorted = efficiency_df.sort_values(by=['Efficiency'], ascending=False)

# Print sorted DataFrame
print(df_sorted)
```
